```python
#importing dataset
import pandas as pd
import numpy as np

customers = pd.read_csv('customers_dataset.csv',
                 dtype={'customer_zip_code_prefix': str},
                 quotechar='"')
print(customers)
```

```python
#checking null values
customers.isnull().sum()
```

```python
#checking colums
print(customers.columns)
```

```python
#deleting spaces and replacing with lower case
customers['customer_city'] = customers['customer_city'].str.lower().str.strip()
customers['customer_state'] = customers['customer_state'].str.upper().str.strip()
print(customers['customer_state'].unique())
```

```python
#checking to duplicates
print(customers.duplicated().sum())
```

```python
#zip_code column is string value and has 5 symbols
customers['customer_zip_code_prefix'] = customers['customer_zip_code_prefix'].astype(str).str.zfill(5)
```

```python
# create new column customer_region
region_map = {
    'SP': 'Sudeste', 'RJ': 'Sudeste', 'MG': 'Sudeste', 'ES': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul',
    'BA': 'Nordeste', 'PE': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste',
    'PA': 'Norte', 'AM': 'Norte', 'AC': 'Norte', 'RO': 'Norte',
    'GO': 'Centro-Oeste', 'DF': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste'}

customers['customer_region'] = customers['customer_state'].map(region_map)
```

```python
customers[['customer_region', 'customer_state']].head(10)
```

```python
#deleting spaces and '' in the rows
customers['customer_city'] = customers['customer_city'].str.replace('"', '', regex=False)
```

```python
#uploading clean dataset
customers.to_csv('customers_clean.csv', index=False, encoding='utf-8-sig')
```

```python
#importing dataset
orders = pd.read_csv('orders_dataset.csv')
orders
```

```python
orders.head()
```

```python
orders.info()
```

```python
orders.isnull().sum()
```

```python
#make sure that colmns with date in type datetime
date_cols = ['order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']
for col in date_cols:
  orders[col] = pd.to_datetime(orders[col], format = '%Y-%m-%d %H:%M:%S', errors = 'coerce')
```

```python
#orders with delivered status has null values
orders[orders['order_status'] == 'delivered'][['order_delivered_customer_date']].isnull().sum()
```

```python
orders.dropna(subset =['order_delivered_customer_date'])
```

```python
# Удаляем заказы, у которых дата одобрения раньше даты покупки
orders = orders[orders['order_approved_at'] >= orders['order_purchase_timestamp']]

# Удаляем заказы, у которых доставка раньше покупки
orders = orders[orders['order_delivered_customer_date'] >= orders['order_purchase_timestamp']]

# Аналогично для передачи перевозчику
orders = orders[orders['order_delivered_carrier_date'] >= orders['order_purchase_timestamp']]
```

```python
#time between
orders['days_to_approve'] = (orders['order_approved_at'] - orders['order_purchase_timestamp']).dt.total_seconds() / 3600 / 24
orders['days_to_carrier'] = (orders['order_delivered_carrier_date'] - orders['order_purchase_timestamp']).dt.total_seconds() / 3600 / 24
orders['days_in_transit'] = (orders['order_delivered_customer_date'] - orders['order_delivered_carrier_date']).dt.total_seconds() / 3600 / 24
orders['total_delivery_time'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.total_seconds() / 3600 / 24

#late and early delivered
orders['delivered_late_days'] = (orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']).dt.total_seconds() / 3600 / 24
orders
```

```python
#max days to deliver 209
print(max(orders['total_delivery_time']))

orders['delivery_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days
print(orders['delivery_days'])

#avg delivery days
avg = orders['delivery_days'].mean()
print(avg)
```

```python
#uploading clean dataset
orders.to_csv('orders_clean.csv', index=False, encoding='utf-8-sig')
```
